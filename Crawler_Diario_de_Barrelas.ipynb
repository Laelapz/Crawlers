```
!apt-get update 
!pip install selenium
!apt install chromium-chromedriver
#Fazendo a instalação do selenium no colab pq não reconheceu no meu pc

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
#Coisas necessárias do Selenium

from bs4 import BeautifulSoup
import requests
#Coisas necessárias do BeautifulSoup

import json
import re

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
wd = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
driver =webdriver.Chrome('chromedriver',chrome_options=chrome_options)
#Configurações do Drive que vai interagir com a página simulando um navegador Chrome

driver.get("https://www.diariodebarrelas.com.br/category/noticias/")
#Acessando a página

cont = 0
while(True):
  try:
    boolean = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, 'cs-load-more')))
    #O drive espera por até no máximo 10s para o botão ser carregado e estar disponível para o clique
    boolean.click()

    print(boolean, cont)
    cont = cont+1
    #Prints para orientação pessoal

  except:
    break
    #Caso o botão não responda em 10s ele impede o erro que pararia a execução e continua o código(!Ou caso tenha chegado ao fim do exibir mais!)

soup = BeautifulSoup(driver.page_source, 'html.parser')
#Coletando o html para o BeautifulSoup

divs = soup.findAll("div", attrs={"class": "cs-posts-area__main  cs-posts-area__archive cs-posts-area__list cs-posts-area__image-width-half cs-display-borders-between-posts cs-posts-area__withsidebar"})
texto = soup.findAll("h2")
#Navegando através dos articles da página

links = []

for div in texto:
  links.append(div.a['href'])
  #Coletando todos os links vinculados as tags <a>
 
dicionario = {}

for link in links:
  url = link
  #Iterando por todos os links coletados para entrar em cada página e pegar o título e texto delas

  headers = {
      'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
  }
  #Simulando um Mozilla, pra não ocorrer erro de autentificação

  print(url)

  response = requests.get(url, headers=headers)
  data = response.text
  soup = BeautifulSoup(data, 'html.parser')
  #Coletando o html para o BeautifulSoup

  title = soup.find('h1', attrs={'class':'cs-entry__title cs-entry__title-line'}).text
  latim = soup.findAll('div',attrs={'class':'entry-content'})
  #Coletando os títulos e os textos

  dicionario.update({"is_sarcastic:": ""})
  dicionario.update({"article_link":url})
  dicionario.update({"headline":title})
  #Adicionando ao dicionário os textos para já ficar no formato antes de salvar no .json
  #O formato será: {"is_sarcastic:": "", "article_link": link, "headline": título, "text": texto}

  print(title)
  for partes in latim:
    partes = partes.text
    
    regex = re.compile(r'[\n\t\r]')
    partes = regex.sub(" ", partes)
    dicionario.update({"text":partes})

  print("-----------"*30)

  with open("sample.json", "a", encoding='utf8') as outfile:    
    json.dump(dicionario, outfile, ensure_ascii=False)
    outfile.write("\n")

```
