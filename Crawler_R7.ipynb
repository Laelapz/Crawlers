```
!apt -get update
!pip install Selenium 
!apt install chromium-chromedriver

from bs4 import BeautifulSoup
#Pega o html da página

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
#Permite a interação com a página alterando o estado dela ou navegando através do site

import requests
#Possibilita fazer a requisição para a página

import json
#Modulo para salvar em .json

import re
#Filtragem dos textos

chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
#Configurações do Driver que vai interagir com a página simulando um navegador Chrome
#É através do drive que o Selenium interage

driver.get("https://www.r7.com/")
#Passando o link com o qual o driver vai interagir
```

```
driver.execute_script("window.scrollTo(0, 55000)")
#Mandando o site rolar até os pixeis pedidos, caso quiser subi é só colocar o valor negativo

soup = BeautifulSoup(driver.page_source, 'html.parser')
#passando o html do drive para a página, também pode ser feito passando uma url

a = soup.findAll('main', attrs={'class':'widget-24x1-ultimas__articles tab_5d8d9785ca908444b7001d21 tab tab--active'})
b = soup.findAll('div', attrs={'class':'widget-24x1-ultimas__img'})
#filtrando através das tags

links = []
for elemento in b:
  if(re.search(r"http:\/\/noticias\.r7\.com.*", elemento.a['href']) != None):
    links.append(elemento.a['href'])
#Passando de notícia em notícia para coletar a url 
    
rpt = 0
dicionario = {}

print(len(links))
#Printando a quantia de noticias que vamos analisar

for link in links:
  url = link

  print(rpt)
  rpt = rpt+1
  #print para poder se orientar quanto falta para acabar a extração dos dados

  headers = {
      'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10.12; rv:55.0) Gecko/20100101 Firefox/55.0',
  }
  #Simulando um Mozilla, pra não ocorrer erro de autentificação

  try:
    response = requests.get(url, headers=headers)
    data = response.text
    soup = BeautifulSoup(data, 'html.parser')
    #Coletando o html para o BeautifulSoup

    
    title = soup.find('h1', attrs={'class':['toolkit-title', 'title text-big']}).text
    #Coletando o titulo que está vinculado a tag <h1>

    latim = soup.findAll('p', attrs={'class':''})
    #Coletando todos os parágrafos vinculados as tags <p>

    title = re.sub("\\s+", " ", title)
    #"\\s+" -> é a regra usada no regex para remover símbolos
    #Que vai ser substituida por um espaço -> " "
    #No texto fornecido, no caso a variável title

    title = re.findall(r"\w.*\w", title)

    

    cont=0
    for partes in latim:
      cont = cont + len(partes.text)
      #Contando a quantia de caracteres pra não pegar textos curtos demais

    text_temp = ""
    for partes in latim:
      partes = partes.text
      text_temp = text_temp+partes
    regex = re.compile(r'[\n\t\r]')
    text_temp = regex.sub(" ", text_temp)
    #Limpando texto de símbolos e juntando todas as linhas que estavam separadas na lista
    

    if(cont>300):
      dicionario.update({"is_sarcastic:": ""})
      dicionario.update({"article_link":url})
      dicionario.update({"headline":title[0]})
      dicionario.update({"text":text_temp})
      #Adicionando ao dicionário os textos para já ficar no formato antes de salvar no .json
      #O formato será: {"is_sarcastic:": "", "article_link": link, "headline": título, "text": texto}

      with open("R7.json", "a", encoding='utf8') as outfile:
        json.dump(dicionario, outfile, ensure_ascii=False)
        outfile.write("\n")
    
  except:
        print("nossa um erro!!") 

```
