```
from bs4 import BeautifulSoup
import requests
import re

url = "https://pt.wikipedia.org/wiki/Filosofia"
response = requests.get(url)
data = response.text
soup = BeautifulSoup(data, 'html.parser')


for sup in soup.find_all("sup", attrs={'class' : 'reference'}):
  sup.decompose()

for span in soup.find_all("span"):
  span.decompose()


print(soup.title.text)
titles = soup.find_all(['p'])

for title in titles:
  print(title.text)
```
